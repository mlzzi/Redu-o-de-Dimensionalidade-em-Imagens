```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
```

## Função para ler imagem a partir de uma URL


```python
def ler_imagem_de_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content)).convert('RGB')  # Converte para RGB
    return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)  # Converte para formato OpenCV (BGR)
```

## Função para converter em tons de cinza

```python
def converter_para_cinza(url):
    # Ler e processar a imagem
    img = ler_imagem_de_url(url)
    img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Exibir a imagem original e em tons de cinza
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(img_cinza, cmap='gray')
    plt.title("Tons de Cinza")
    plt.axis("off")

    plt.show()
```

## Função para converter em preto e branco

```python
def converter_para_preto_branco(url, limiar=160):
    # Ler e processar a imagem
    img = ler_imagem_de_url(url)
    img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    suave = cv2.GaussianBlur(img_cinza, (7, 7), 0)  # Aplicar suavização
    _, binarizada = cv2.threshold(suave, limiar, 255, cv2.THRESH_BINARY)

    # Exibir a imagem original e binarizada
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(binarizada, cmap='gray')
    plt.title("Preto e Branco")
    plt.axis("off")

    plt.show()
```

## Variáveis para guardar URL das imagens

```python
# URL das imagens
url_imagem1 = "https://www.midiorama.com/wp-content/uploads/2019/04/lena-meyer-landrut.png"

url_imagem2 = "https://cdn.motor1.com/images/mgl/lEmjGg/s1/chevrolet-tracker-rs-2024.webp"
```

## Execução dos códigos

```python
# Chamando as funções
converter_para_cinza(url_imagem1)
converter_para_preto_branco(url_imagem1)

converter_para_cinza(url_imagem2)
converter_para_preto_branco(url_imagem2)
```
